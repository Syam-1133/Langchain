converting text into vectors

```python
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
os.environ['OPENAI_API_KEY'] = os.getenv("OPENAI_API_KEY")
```

```python
from langchain.embeddings import OpenAIEmbeddings
embedings =OpenAIEmbeddings(model="text-embedding-3-large")
embedings
```

```python
text = " This is a test sentence."
query_results = embedings.embed_query(text)
query_results
```

```python
len(query_results)
```

```python
embedings_1024 = OpenAIEmbeddings(model="text-embedding-3-large", dimensions=1024)
```

```python
embedings_1024
```

```python
text = " This is a test sentence."
quary_results = embedings_1024.embed_query(text)
len(quary_results)
```

```python
quary_results
```

```python
from langchain.document_loaders import TextLoader
loader = TextLoader("speech.txt")
docs = loader.load()
docs
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter=RecursiveCharacterTextSplitter(chunk_size=500,chunk_overlap=50)
final_documents=text_splitter.split_documents(docs)
final_documents
```

```python
## Vector Embedding And Vector StoreDB

from langchain.vectorstores import Chroma
db = Chroma.from_documents(final_documents,embedings_1024)
db
```

```python
### Retrieve the results from query vectorstore db
query="It will be all the easier for us to conduct ourselves as belligerents"
retrieved_results=db.similarity_search(query)
print(retrieved_results)
```
